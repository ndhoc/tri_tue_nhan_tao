```python
import random

def rule(x, y):
  moves=[]
  if(x>0): moves.append((x-1, y))
  if(x<2): moves.append((x+1, y))
  if(y>0): moves.append((x, y-1))
  if(y<2): moves.append((x, y+1))
  return moves

persistent_state=[]
persistent_model=[]
persistent_rules=rule
persistent_action=""

duong_di=[];lich_su=[]

def action(a, x, y):
    global persistent_state,persistent_model,persistent_action
    duong_di.append((x, y));lich_su.append((x, y))
    persistent_state = a;persistent_model = lich_su

    if(a[x][y]==1):
        a[x][y] = 0
        hanh_dong = "DON"
        persistent_action = hanh_dong
        return a, x, y, hanh_dong

    moves=persistent_rules(x, y);hop_le=[]
    for dx,dy in moves:
        if((dx,dy) not in persistent_model):
            hop_le.append((dx,dy))

    if(len(hop_le)==0): hop_le=moves
    dx, dy = random.choice(hop_le)
    hanh_dong = "DI" + str((dx, dy))
    persistent_action = hanh_dong
    return a,dx,dy,hanh_dong

a=[];x=random.randint(0,2);y=random.randint(0,2)

for i in range(3):
    row = []
    for j in range(3):
        row.append(random.randint(0, 1))
    a.append(row)

for row in a:print(row)
print("Vi tri bat dau:", (x,y))

t = 10
while(t > 0):
    a,x,y,hanh_dong = action(a, x, y)
    t=t-1
print("Lich su:")
for i in lich_su:print(i, end=" -> ")
print()
print("Ket qua:")
for i in a:print(i)
```
