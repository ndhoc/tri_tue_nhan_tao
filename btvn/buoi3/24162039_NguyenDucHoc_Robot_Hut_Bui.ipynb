```python
import random

N = 5; SACH = 0; BAN = 1; VAT_CAN = -1
matrix = [[random.choice([0, 1, -1]) for _ in range(N)] for _ in range(N)]
matrix[0][0] = 0
x, y = 0, 0
prev_x, prev_y = 0, 0
model = [[None] * N for _ in range(N)]
action = None

for row in matrix: print(row)
for step in range(10):
    percept = matrix[x][y]
    prev_x, prev_y = x, y
    model[x][y] = percept
    if percept == BAN: rule_action = "HUT_BUI"
    else: rule_action = "DI_CHUYEN"
    if rule_action == "HUT_BUI":
        print(f"Lan {step}: Co rac o ({x}, {y}) -> Hut bui.")
        matrix[x][y] = SACH
        model[x][y] = SACH
    elif rule_action == "DI_CHUYEN":
        huong_di = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        random.shuffle(huong_di)

        for dx, dy in huong_di:
            nx, ny = x + dx, y + dy
            if 0 <= nx < N and 0 <= ny < N and matrix[nx][ny] != VAT_CAN:
                x, y = nx, ny
                break
        print(f"Lan {step}: O sach, di chuyen tu ({prev_x}, {prev_y}) den ({x}, {y})")
```
